Problem1:  Getting the count of Movies having Average rating greater than 4.5

```python
import pandas as pd

unames = ['user_id', 'gender', 'age', 'occupation', 'zip']
users = pd.read_table('C:/Users/bkafley/Desktop/stuff/DS 501/Case2/ml-1m/users.dat', sep='::', header=None,
names=unames)
rnames = ['user_id', 'movie_id', 'rating', 'timestamp']
ratings = pd.read_table('C:/Users/bkafley/Desktop/stuff/DS 501/Case2/ml-1m/ratings.dat', sep='::', header=None,
names=rnames)
mnames = ['movie_id', 'title', 'genres']
movies = pd.read_table('C:/Users/bkafley/Desktop/stuff/DS 501/Case2/ml-1m/movies.dat', sep='::', header=None,
names=mnames)
movies
```

I have used here np.average to calcualte average ratings, but we can use aggfunc='mean' here too.

```python
data = pd.merge(pd.merge(ratings, users), movies)
Data_WH = pd.HDFStore('C:/Users/bkafley/Desktop/stuff/DS 501/Case2/ml-1m/CaseStudy2.h5')
Data_WH['data1'] = data
Data_WH.close()
avg_overall_ratings = data.pivot_table('rating', index= 'title',aggfunc= np.average)
di2= dict(avg_overall_ratings)
di2_45 = dict((k,v) for k,v in di2.items() if v >=4.5)
print 'Movies having average ratings greater than 4.5 = ' + str (len(di2_45))
```

#Part iv
We define the concept of Popularity as many users showing their interest to rate the movies.
Since every movies has atleast one user rating and the max rating for any movie is 3428. We find all the movies which have users atleast greater than the median number of the users.
We use median here as we feel that this is the skewed data.(One userid may have voted for 90% of the movies with high rating)
create series of each column with all statistics and then create a new data frame by joining them.

```python
title_group= data.groupby('title')
title_grp_data = title_group.user_id.count()
mean_users = title_grp_data.median()
mean_users
```

Series of user ids more than median

```python
mov_over_meadian_users = title_grp_data[title_grp_data > mean_users]
mov_over_meadian_users[:5]
```

Mean Rating for all movies

```python
mean_ratings = title_group.rating.mean()
mean_ratings[:5]
```

```python
merged = pd.concat([mov_over_meadian_users,mean_ratings], axis =1).reset_index()
```

```python
merged.rename(columns = {'index':'title'}, inplace=True)
merged[:5]
```

We can also replace the NAN values but we filter on user_id value greater than 0. These are popular ten movies with good number of users and mean ratings more than 4.4 since 4.5 gives less than 10 movies.

```python
top_popular_movies = merged.loc[(merged['user_id'].values >0) & (merged['rating'] > 4.4)]
tten = top_popular_movies.sort('rating', ascending=False)
tten[:10]
```

If we donot use the critereia of many users voting for movies and just use the mean ratings we can get different results.(Which have avearge rating of 5 but might not have sufficient users voting)

```python
top_popular_movies2 = merged.loc[(merged['rating'] > 4.5)]
tten2 = top_popular_movies2.sort('rating', ascending=False)
tten2[:10]
```
